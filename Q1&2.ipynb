### Q1)1)  
In theory we know:  
### poisson distribution :     MEAN=lambda=10                 VARIANCE=lambda=10  
### expenential distribution : MEAN=1/(lambda)=2          VARIANCE=1/(lambda^2)=4  
### geometric distribution :   MEAN=(1-p)/p=4               VARIANCE=(1-p)/p^2=20  


```python
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd
np.random.seed(545)
```

```python
#PART 2&3)
lambda_poisson = 10
lambda_exponential = 0.5 
p_geometric = 0.2  


sample_sizes = [30, 300, 3000]
num_samples = 1000

samples_poisson = {size: [] for size in sample_sizes}
samples_exponential = {size: [] for size in sample_sizes}
samples_geometric = {size: [] for size in sample_sizes}
mean_poisson={size:[] for size in sample_sizes}
mean_exponential={size:[] for size in sample_sizes}
mean_geometric={size:[] for size in sample_sizes}


for size in sample_sizes:
    for i in range(num_samples):
        #samples
        samples_poisson[size].append(np.random.poisson(lambda_poisson, size))
        samples_exponential[size].append(np.random.exponential(1/lambda_exponential, size))
        samples_geometric[size].append(np.random.geometric(p_geometric, size))
        #means
        mean_poisson[size].append(np.mean(np.random.poisson(lambda_poisson, size)))
        mean_exponential[size].append(np.mean(np.random.exponential(1/lambda_exponential, size)))
        mean_geometric[size].append(np.mean(np.random.geometric(p_geometric, size)))

poisson_range = (min(min(mean_poisson[size]) for size in sample_sizes), max(max(mean_poisson[size]) for size in sample_sizes))
exponential_range = (min(min(mean_exponential[size]) for size in sample_sizes), max(max(mean_exponential[size]) for size in sample_sizes))
geometric_range = (min(min(mean_geometric[size]) for size in sample_sizes), max(max(mean_geometric[size]) for size in sample_sizes))

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sn.histplot(mean_poisson[30],binrange=poisson_range)
plt.title('Poisson Distribution 30')
plt.text(0.95, 0.95, f'Mean: {np.mean(mean_poisson[30]):.4f}\nStd: {np.std(mean_poisson[30]):.4f}', 
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

plt.subplot(1, 3, 2)
sn.histplot(mean_exponential[30],binrange=exponential_range)
plt.title('Exponential Distribution 30')
plt.text(0.95, 0.95, f'Mean: {np.mean(mean_exponential[30]):.4f}\nStd: {np.std(mean_exponential[30]):.4f}', 
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

plt.subplot(1, 3, 3)
sn.histplot(mean_geometric[30],binrange=geometric_range)
plt.title('Geometric Distribution 30')
plt.text(0.95, 0.95, f'Mean: {np.mean(mean_geometric[30]):.4f}\nStd: {np.std(mean_geometric[30]):.4f}', 
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sn.histplot(mean_poisson[300],binrange=poisson_range)
plt.title('Poisson Distribution 300')
plt.text(0.95, 0.95, f'Mean: {np.mean(mean_poisson[300]):.4f}\nStd: {np.std(mean_poisson[300]):.4f}', 
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

plt.subplot(1, 3, 2)
sn.histplot(mean_exponential[300],binrange=exponential_range)
plt.title('Exponential Distribution 300')
plt.text(0.95, 0.95, f'Mean: {np.mean(mean_exponential[300]):.4f}\nStd: {np.std(mean_exponential[300]):.4f}', 
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

plt.subplot(1, 3, 3)
sn.histplot(mean_geometric[300],binrange=geometric_range)
plt.title('Geometric Distribution 300')
plt.text(0.95, 0.95, f'Mean: {np.mean(mean_geometric[300]):.4f}\nStd: {np.std(mean_geometric[300]):.4f}', 
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sn.histplot(mean_poisson[3000],binrange=poisson_range)
plt.title('Poisson Distribution 3000')
plt.text(0.95, 0.95, f'Mean: {np.mean(mean_poisson[3000]):.4f}\nStd: {np.std(mean_poisson[3000]):.4f}', 
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

plt.subplot(1, 3, 2)
sn.histplot(mean_exponential[3000],binrange=exponential_range)
plt.title('Exponential Distribution 3000')
plt.text(0.95, 0.95, f'Mean: {np.mean(mean_exponential[3000]):.4f}\nStd: {np.std(mean_exponential[3000]):.4f}', 
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

plt.subplot(1, 3, 3)
sn.histplot(mean_geometric[3000],binrange=geometric_range)
plt.title('Geometric Distribution 3000')
plt.text(0.95, 0.95, f'Mean: {np.mean(mean_exponential[3000]):.4f}\nStd: {np.std(mean_exponential[3000]):.4f}', 
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

plt.tight_layout()
plt.show()
```

### 4)  
According to central limit theory ("قضیه حد مرکزی") sum of many independent variables with a finite mean and variance as the number of variables increases , shapes a normal disribution .  
In Poisson distribution with parameter ("λ"), Using CLT says the sum of the average(mean) increses and forms a normal distribution as we can see in the plots.  
The same goes for other distributions aswell.  
The number of triales(sample size) and mean and std are the parameters forming this normal distribution.

```python
samples_nor_poisson = {size: [] for size in sample_sizes}
samples_nor_exponential = {size: [] for size in sample_sizes}
samples_nor_geometric = {size: [] for size in sample_sizes}
mean_nor_poisson={size:[] for size in sample_sizes}
mean_nor_exponential={size:[] for size in sample_sizes}
mean_nor_geometric={size:[] for size in sample_sizes}

for size in sample_sizes:
    for i in range(num_samples):
        #samples
        samples_nor_poisson[size].append(np.random.normal(10,10 ** 0.5,size))
        samples_nor_exponential[size].append(np.random.normal(2,2,size))
        samples_nor_geometric[size].append(np.random.normal(4,20 ** 0.5,size))
        #means
        mean_nor_poisson[size].append(np.mean(samples_nor_poisson[size][-1]))
        mean_nor_exponential[size].append(np.mean(samples_nor_exponential[size][-1]))
        mean_nor_geometric[size].append(np.mean(samples_nor_geometric[size][-1]))


plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sn.histplot(mean_nor_poisson[30],binrange=poisson_range)
plt.title('Normal Distribution from Poisson Distribution 30')
plt.text(0.95, 0.95, f'Mean: {np.mean(mean_nor_poisson[30]):.4f}\nStd: {np.std(mean_nor_poisson[30]):.4f}', 
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

plt.subplot(1, 3, 2)
sn.histplot(mean_nor_exponential[30],binrange=exponential_range)
plt.title('Normal Distribution from Exponential Distribution 30')
plt.text(0.95, 0.95, f'Mean: {np.mean(mean_nor_exponential[30]):.4f}\nStd: {np.std(mean_nor_exponential[30]):.4f}', 
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

plt.subplot(1, 3, 3)
sn.histplot(mean_nor_geometric[30],binrange=geometric_range)
plt.title('Normal Distribution from Geometric Distribution 30')
plt.text(0.95, 0.95, f'Mean: {np.mean(mean_nor_geometric[30]):.4f}\nStd: {np.std(mean_nor_geometric[30]):.4f}', 
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sn.histplot(mean_nor_poisson[300],binrange=poisson_range)
plt.title('Normal Distribution from Poisson Distribution 300')
plt.text(0.95, 0.95, f'Mean: {np.mean(mean_nor_poisson[300]):.4f}\nStd: {np.std(mean_nor_poisson[300]):.4f}', 
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

plt.subplot(1, 3, 2)
sn.histplot(mean_nor_exponential[300],binrange=exponential_range)
plt.title('Normal Distribution from Exponential Distribution 300')
plt.text(0.95, 0.95, f'Mean: {np.mean(mean_nor_exponential[300]):.4f}\nStd: {np.std(mean_nor_exponential[300]):.4f}', 
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

plt.subplot(1, 3, 3)
sn.histplot(mean_nor_geometric[300],binrange=geometric_range)
plt.title('Normal Distribution from Geometric Distribution 300')
plt.text(0.95, 0.95, f'Mean: {np.mean(mean_nor_geometric[300]):.4f}\nStd: {np.std(mean_nor_geometric[300]):.4f}', 
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sn.histplot(mean_nor_poisson[3000],binrange=poisson_range)
plt.title('Normal Distribution from Poisson Distribution 3000')
plt.text(0.95, 0.95, f'Mean: {np.mean(mean_nor_poisson[3000]):.4f}\nStd: {np.std(mean_nor_poisson[3000]):.4f}', 
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

plt.subplot(1, 3, 2)
sn.histplot(mean_nor_exponential[3000],binrange=exponential_range)
plt.title('Normal Distribution from Exponential Distribution 3000')
plt.text(0.95, 0.95, f'Mean: {np.mean(mean_nor_exponential[3000]):.4f}\nStd: {np.std(mean_nor_exponential[3000]):.4f}', 
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

plt.subplot(1, 3, 3)
sn.histplot(mean_nor_geometric[3000],binrange=geometric_range)
plt.title('Normal Distribution from Geometric Distribution 3000')
plt.text(0.95, 0.95, f'Mean: {np.mean(mean_nor_exponential[3000]):.4f}\nStd: {np.std(mean_nor_exponential[3000]):.4f}', 
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes)

plt.tight_layout()
plt.show()
```
